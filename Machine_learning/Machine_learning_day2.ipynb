분류

이진 분류기: 데이터를 두 그룹으로 분류하는 것.

Confusion matrix (오차행렬)

precision(정밀도): 양성(positive)이라 예측한 것 중 진짜 양성

recall(재현율): 실제 양성 중 양성이라 잘 예측한 비율
→ 실제 값이 중요한 업계에서 중요하게 여겨짐.
→ 모델 사용 목적에 따라 중요도 달라짐.

F1 점수: 정밀도와 재현율의 조화 평균. 정밀도와 재현율 둘 다 보는 효과를 얻을 수 있음. 
→ cut off(결정 임계값, decision threshold) 엄격하게 잡으면 재현율 떨어짐.

ROC 곡선: 거짓 양성 비율에 대한 참 양성 비율(재현율)의 관계 나타냄. 

false 개수의 영향을 많이 받는 지표. 

가운데에 곡선이 그려질 수록 예측이 낮은 것. 

왼쪽으로 치우친 곡선(정확한 예측), 오른쪽으로 치우친 곡선(반대로 예측)이 좋음.

AUC: ROC 커브가 y축에 최대한 근접해야 하고, ROC 커브 아래 면적, 즉 AUC가 1에 가까울 수록 좋은 성능임을 나타냄
다중 클래스 분류(다항 분류): 세 개 이상의 그룹 분류 과정.

One versus all: 타겟 데이터랑 나머지 모든 데이터를 분류해내는 모델. 특정 클래스만 독립적 구분.

One versus one: 타겟 데이터와 다른 한 데이터 분류해내는 모델. 특정 클래스와 다른 클래스만 구분.

분류 알고리즘1

결정트리: 데이터에 있는 규칙을 학습을 통해 자동으로 찾아내 트리(Tree) 기반의 분류 규칙을 만듦. (If-Else 기반 규칙) 기계가 자동 학습.

max_depth: 분류 기준을 몇 개 까지 볼 것인가에 대한 것.

gini: 불순도. 불순도 높으면 다른 클래스가 혼재되어 있음 의미. 0이라면 한 종류의 클래스만 있다는 의미.

samples: 전체 n 수
value: 클래스별 n수

class: 각 클래스별 비율을 계산하여 가장 높은 비율에 해당하는 클래스 선정

선을 계속 그어 클래스를 구분하는 방법을 채택하고 있기에 과적합에 취약.

규제 하이퍼파라미터

max_depth  2) min_samples_split  3) min_samples_leaf
		      4)  min_wight_fractio_leaf  5) max_leaf_nodes  6) max_features

KNN 알고리즘: 주변 데이터(neighbor)를 살펴본 뒤 더 많은 데이터가 포함되어 있는 클래스로 분류하는 방식 k개수만큼 주변 데이터 살핌. 비지도 학습!

유클리드 거리: 차원 수를 늘려줌.

GMM 클러스터링 기법: 장점) 실제 분류 성능이 좋음  단점) 분포 간 경계에 위치해 있는 값의 경우 명확한 분류가 어려움.

분류 알고리즘2

편향: 예측값과 실제값이 얼마나 차이를 보이느냐의 정도. 편향이 크다면 원하는 분류 결과가 나오기 어려움. why? 목표 값에서 떨어져 있기 때문.

분산: 한 그룹 내의 값이 얼마나 퍼져있는지 모여있는지의 정도. 분산이 크다면 일정한 결과를 얻기 어려움.분산이 작다면 오버피팅 의심 가능성이 높아질 수 있음. 
→ 편향, 분산 둘 다 적어야 함!

앙상블 모델: 여러 분류기 생성하고 돌림. 각 분류 결과를 모아 공통적으로 나타는 분류 경향에 따라 분류 결과를 결정하는 모델.
 
보팅: 하드 보팅-다수 분류기가 예측한 결과값을 최종 결과로 선정

         소프트 보팅-모든 분류기가 예측한 레이블 값의 결정 확률 평균을 구한 뒤 가장 확률이 높은 레이블 값을 최종 결과로 선정

배깅: 훈련 세트를 각 모델에 학습. 

그 결과 가장 많이 훈련 했던 클래스를 선정(하드 보팅). 하드 보팅 + 모두 같은 유형의 알고리즘 기반의 분류기 사용
→ 랜덤 포레스트 알고리즘: 의사결정 나무 여러개 만들어서 배깅 실시하는 것. 

/ 데이터를 샘플링 하게 되면 표본의 다양성을 많이 추가하게 되기 때문에, 분산이 줄어드는 효과를 가져옴. 결정경계의 불규칙성 감소. 일반화 성능 증가.

부스팅: 여러 개 분류기(classifier)가 순차적 학습. 

why 순차적 학습? 데이터 편향 시 기계가 스스로 다음 분류기에게 가중치 부여하며 학습/예측할 수 있게 하기 위해. 최적의 값 찾을 때까지 반복.

 장점) 피쳐 엔지니어링을 기계가 스스로 함. 단점) 피쳐 엔지니어링이 학습 후 실시되기에 훈련시간 비례 증가.

 
→ 에이다부스트: 하나의 모델을 훈련시킨 후 잘 못 예측된 샘플을 보다 강조하면서 해당모델을 다시 훈련. 가중치0, 입력 데이터 유지.
→ 그래디언트 부스팅: 에이다부스트 기법은 샘플의 가중치를 조정하는 반면, 그래디언트 부스팅기법은 이전 예측기에 의해 생성된 잔차(residual error)에 대해 새로운 예측기를 학습하는 방식으로 모델 조절. 가중치x, 입력 데이터가 바뀜.
스태킹: 배깅과 유사. 여러 예측값을 학습데이터로 활용하는 예측기를 학습.


```python
# 패키지 및 데이터 불러오기
import seaborn as sns
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=2)
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# PCA 차원축소
iris_pca = pca.fit_transform(iris.iloc[:,:-1])
iris_pca_df = pd.DataFrame(iris_pca, columns=['PCA_component_1', 'PCA_component_2'])

# 시각화
sns.scatterplot(data=iris_pca_df, x='PCA_component_1', y='PCA_component_2')
```

```python
# GMM 클러스터링
gmm = GaussianMixture(n_components=3, random_state=12345)
gmm_y = gmm.fit_predict(iris_pca_df)

# GMM 클러스터링 결과 시각화
iris_pca_df['y'] = gmm_y
sns.scatterplot(data=iris_pca_df, x='PCA_component_1', y='PCA_component_2', hue='y')
plt.show()
```

```python
# species 정보 시각화
iris_pca_df['ground_truth'] = iris['species']
sns.scatterplot(data=iris_pca_df, x='PCA_component_1', y='PCA_component_2', hue='ground_truth')
plt.show()
```

```python
# 모델 적용: 배깅
from sklearn.linear_model import LogisticRegression  # 로지스틱 회귀 모델
from sklearn.ensemble import BaggingClassifier # BaggingClassifier 불러오기
# 성능 평가를 위한 지표 불러오기
from sklearn.metrics import accuracy_score, recall_score, precision_score, roc_auc_score, confusion_matrix
lr_clf = LogisticRegression(solver='liblinear') # 로지스틱 회귀 모델 생성

bagging_clf = BaggingClassifier(estimator=lr_clf) # BaggingClassifier 모델 생성

# BaggingClassifier 학습/예측.
bagging_clf.fit(X_train , y_train)
pred = bagging_clf.predict(X_test)

# 평가

# 정확도 (Accuracy) 계산
accuracy = accuracy_score(y_test, pred)
# 재현율 (Recall) 계산
recall = recall_score(y_test, pred)
# 정밀도 (Precision) 계산
precision = precision_score(y_test, pred)
# ROC AUC 점수 계산
auc = roc_auc_score(y_test, pred)
# 혼동 행렬 (Confusion Matrix) 계산
matrix = confusion_matrix(y_test, pred)

print('Bagging 분류기 정확도: {0:.4f}'.format(accuracy))
print('Bagging 분류기 Recall: {0:.4f}'.format(recall))
print('Bagging 분류기 Precision: {0:.4f}'.format(precision))
print('Bagging 분류기 AUC: {0:.4f}'.format(auc))
print('Bagging 분류기 Confusion Matrix:','\n', matrix)
```

```python
from sklearn.metrics import roc_curve, precision_recall_curve
import matplotlib.pyplot as plt

# 예측 확률을 계산
proba = lr_clf.predict_proba(X_test)[:, 1]

# ROC 곡선 계산
fpr, tpr, thresholds = roc_curve(y_test, proba)

# Precision-Recall 곡선 계산
precision, recall, thresholds = precision_recall_curve(y_test, proba)

# ROC 곡선 그리기
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc)
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()

# Precision-Recall 곡선 그리기
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.show()
```
