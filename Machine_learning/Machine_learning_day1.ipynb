머신러닝의 종류
1) 분류: 기준에 따라 데이터 분류.
2) 회귀: 연속적 숫자를 y값으로 두고 예측하는 것. 숫자로 나타낼 수 있는 데이터 예측
3) 군집화: 학습 과정이 분류, 회귀와 다름. 분류, 회귀는 학습 데이터+답의 형태로 학습 실시. 지도 학습! but, 군집화는 답 없이 학습 데이터만 제공하여 스스로 학습하게 함. 비지도 학습!

학습 데이터: 학습을 하기 위해 필요한 데이터 / 테스트 데이터: 모델을 테스트 하기 위해 학습 과정에는 참여하지 않는 데이터. / + 검증 데이터: 과적합 등의 문제를 방지하기 위해 사용.

준지도학습: 레이블이 있는 것, 없는 것을 모두 학습에 사용(레이블=y값). 
자기지도 학습: 레이블이 없는 샘플로 구성된 데이터셋으로부터 레이블을 모두 갖는 샘플로 구성된 데이터셋을 생성하여 학습을 진행하는 기법.
강화 학습: 학습 시스템에서 판단에 따라 가중치를 다르게 주는 것(상/벌).

스케일링
값의 범위를 제한하는 과정
1) standard scaling: 평균0, 분산 1인 정규분포를 나타내도록 스케일링. 데이터의 범위는 달라지지x, 데이터의 분포 달라짐0. 장점 =정규분포에 가까워짐. 모수적인 방법 활용 가능해짐. 머신러닝시 신뢰도 상승.
2) Minmax scaling: 데이터를 0과 1사이의 값으로 축소. 데이터의 분포유지, 데이터의 범위 0~1 사이로 변화. 장점 = 데이터의 범위가 컨트롤 가능한 범위 안으로 들어가게 되는 효과가 있음. 
+ 결측치 처리 1) 제거: 너무 많을 때, 지워도 무의미할 때.  2) 채워주기: 평균, 최빈값
+ 이상치 제거 1) 상한값/하한값으로 대체

모델 선택과 훈련

교차검증: k-fold

학습중인 모델의 성능을 평가하는 방법으로 고정된 검증 데이터셋에만 과적합 되어버리지 않도록 하는 방법

why? 검증 데이터셋에 과적합이 될 수 있으므로 이를 막기 위해 동일 검증 데이터셋이 아닌 매번 다른 검증 데이터셋 사용.

학습 데이터셋을 k개 부분집합으로 무작위 분할 후 학습 k번 반복. 매 학습 시 다른 검증 데이터셋 사용. 최종평가는 k번 평가 결과의 평균값 활용.
장점) 검증 정밀 가능. 단점) k개수만큼 학습 시간이 증가.
데이터를 잘 섞지 않았을 경우 검증이 잘 이루어지지 않는 문제 발생
 → Stratified k-fold으로 해결

교차검증: Stratified k-fold

각 폴드의 학습데이터셋과 검증데이터셋의 레이블 분포가 유사하도록 할당 → 비율을 맞추어 검증 데이터셋 할당.

현업에서 잘 사용x → k-fold사용해서 데이터 잘 섞는것이 더 선호됨, 편향문제가 근본적으로 해결되지 않으므로 k-fold 더 흔히 사용.

+ 다수 범주(majority class) / 소수 범주 (minority class)

```python
df = pd.read_csv('.csv', encoding='cp949')
```

```python
# 사용하지 않는 열 drop
df.drop(['', '', ''],  axis=1, inplace=True)
```

```python
# 결측치 처리 --> N,N인 경우만 N으로 처리 그 외 Y로 처리

# 조건에 따라 '' 열 설정
def set_yn(row):
    if pd.isna(row['']):
        if row[''] == 'N' and row[''] == 'N':
            return 'N'
        else:
            return 'Y'
    else:
        return row['']

# '' 열 설정
df[''] = df.apply(set_yn, axis=1)
```

```python
# 결측치 확인
null_cnt_df = pd.DataFrame(df.isnull().sum()).rename(columns={0: 'null_count'}).reset_index()
null_cnt_df['null_ratio'] = round(null_cnt_df['null_count'] / len(df) *100,2)
null_cnt_df
```

```python
from sklearn.impute import SimpleImputer
import pandas as pd

# SimpleImputer 객체 생성 (strategy='median'으로 중앙값으로 대체 설정)
imputer = SimpleImputer(strategy='median')

# 결측치 대체할 열 선택
cols_to_impute = ['', '','', '']

# 중앙값으로 결측치 대체
df = df.copy()  # 원본 데이터프레임 복사
df[cols_to_impute] = imputer.fit_transform(df[cols_to_impute])

print("Imputed DataFrame:")
print(df)
```

```python
# 결측치 확인
null_cnt_df = pd.DataFrame(df.isnull().sum()).rename(columns={0: 'null_count'}).reset_index()
null_cnt_df['null_ratio'] = round(null_cnt_df['null_count'] / len(df) *100,2)
```

```python
# 결측치 비율이 모두 1퍼센트 미만이므로 행 삭제

# 선택된 열들에 대해 결측치가 있는 행 삭제하여 새로운 데이터프레임 생성
cleaned_df = df.dropna(subset=[])

# 새로운 데이터프레임을 기존의 df에 할당
df = cleaned_df.copy()
```

```python
# 사분위수 계산하여 사분쉬우 기반으로 이상치 선택 및 중앙값 대체

# 사분위수 계산
Q1 = df[''].quantile(0.25)
Q3 = df[''].quantile(0.75)
IQR = Q3 - Q1

# 사분위수를 기반으로 이상치 선택
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

selected_outliers = df[(df[''] < lower_bound) | (df[''] > upper_bound)]
selected_outliers 

# 이상치를 중앙값으로 대체
median_value = df[''].median()
df.loc[(df[''] < lower_bound) | (df[''] > upper_bound), ''] = median_value
```

# 의사결정나무

의사결정나무 모델
다양한 if-else 조건문이 마치 나무처럼 배치되어 실제 데이터의 분류 기준을 정의할 수 있도록 구성하는 방식
classifier → 분류, regression → 예측
+ 화이트 박스 모델: 추상화가 되어 어떻게 동작하는지 모르는것(블랙 박스). 어떻게 동작하는지 구체적으로 뜯어보는 것.
[장점]
사람이 이해하기 간단하며, 모델 시각화0. 화이트 박스 모델을 적용

데이터 전처리를 거의 혹은 전혀 요구하지 않습니다.

모델의 복잡도는 데이터의 복잡도에 비례, 학습에 소요되는 자원(시간, 메모리 등)을 예측하기 쉽습니다.

숫자형 및 범주형 데이터를 가리지 않습니다.
[단점]
과적합에 취약.
입력 데이터의 차이가 조금만 나더라도 트리가 크게 달라질 수 있습니다.

외삽(입력 값의 범위 바깥에서 추정하는 작업)에 적합한 모델은 아닙니다.

최적의 의사결정나무 모델 학습은 NP-완전 문제입니다. 즉, 최적의 알고리즘이 발견되지 않은 상황입니다.

NP-불완전 문제: 모든 최적 해를 구해놓은 문제

```python
# import package
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# fit-predict!
iris_data = load_iris()
X_train , X_test , y_train , y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.2,  random_state=12345)

# DecisionTree Classifier
dt_clf = DecisionTreeClassifier(random_state=12345)
dt_clf.fit(X_train, y_train)

y_pred = dt_clf.predict(X_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
matrix = metrics.confusion_matrix(y_test, y_pred)

print(f'Decision Tree 정확도: {accuracy:.2%}\n')
print('Decision Tree Confusion Matrix:\n', matrix)
```

```python
plot_tree(dt_clf)
plt.show()
```

```python
# feature importance 확인
for name, value in zip(iris_data.feature_names , dt_clf.feature_importances_):
    print(f'{name}: {value:.3f}')

print('-'*50)

# feature importance 시각화
ax = sns.barplot(x=dt_clf.feature_importances_ , y=iris_data.feature_names)
ax.bar_label(ax.containers[0], fmt='  %.3f')
plt.gca().spines[['right', 'top']].set_visible(False)
plt.title('Feature Importances')
plt.show()
```

# k-Fold Cross Validation

```python
# import package
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np
from sklearn.datasets import load_iris
```

```python
# input data, ML model
iris = load_iris()
X = iris.data
y = iris.target
dt_clf = DecisionTreeClassifier(random_state=12345)

# 5-fold CV
kfold = KFold(n_splits=5, shuffle=False)
cv_index = kfold.split(X)

cv_accuracy = []
ind = 1

for train_index, test_index in cv_index:
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    dt_clf.fit(X_train , y_train)
    accuracy = dt_clf.score(X_test, y_test)

    print(f'{ind}번째 Cross Validation 정확도: {accuracy:.2%}')
    cv_accuracy.append(accuracy)

    ind += 1

print(f'''-------------------------------------------
Cross Validation 정확도 평균: {np.mean(cv_accuracy):.2%}''')
```

```python
# import package
import pandas as pd
from sklearn.model_selection import StratifiedKFold
```

```python
# imput data
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# 3-fold CV
kfold = KFold(n_splits=3, shuffle=False)

ind = 1

for train_index, test_index in kfold.split(X):
    train_label = iris_df.loc[train_index, 'species']
    test_label = iris_df.loc[test_index, 'species']
    print(
f'''\033[1m[{ind}번째 3-Fold Cross Validation]\033[0m
* training set 데이터 분포
{train_label.value_counts()}
* test set 데이터 분포
{test_label.value_counts()}'''
    )
    ind += 1

print('-'*50)

# Stratified 3-fold CV
skfold = StratifiedKFold(n_splits=3)

ind = 1

for train_index, test_index in skfold.split(X, y):
    train_label = iris_df.loc[train_index, 'species']
    test_label = iris_df.loc[test_index, 'species']
    print(
f'''\033[1m[{ind}번째 Stratified 3-Fold Cross Validation]\033[0m
* training set 데이터 분포
{train_label.value_counts()}
* test set 데이터 분포
{test_label.value_counts()}'''
    )
    ind += 1
```

# 불균형 데이터를 처리하는 imbalanced-learn

```python
# import package
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from collections import Counter
from matplotlib import pyplot as plt
import seaborn as sns

from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE

X, y = make_classification(n_samples=10000, n_features=5, weights=[0.99], flip_y=0, random_state=1234)
print(Counter(y))
```

```python
# random undersampling
undersample = RandomUnderSampler(random_state=12345)
X_undersampled, y_undersampled = undersample.fit_resample(X, y)
print(Counter(y_undersampled))
```

```python
# random oversampling
oversample = RandomOverSampler(random_state=12345)
X_oversampled, y_oversampled = oversample.fit_resample(X, y)
print(Counter(y_oversampled))
```

```python
# random oversampling, sampling_strategy=0.5
oversample = RandomOverSampler(random_state=12345, sampling_strategy=0.5)
X_resampled, y_resampled = oversample.fit_resample(X, y)
print(Counter(y_resampled))
```

```python
# SMOTE oversampling
smote_sample = SMOTE(random_state=12345, sampling_strategy=0.5)
X_smote, y_smote = smote_sample.fit_resample(X, y)
print(Counter(y_smote))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 15))
sns.scatterplot(x=X[:, 1], y=X[:, 2], hue=y, ax=axes[0][0])
axes[0][0].set_title('Original Data')
sns.scatterplot(x=X_undersampled[:, 1], y=X_undersampled[:,2], hue=y_undersampled, ax=axes[0][1])
axes[0][1].set_title('Random Undersampling')
sns.scatterplot(x=X_oversampled[:, 1], y=X_oversampled[:, 2], hue=y_oversampled, ax=axes[1][0])
axes[1][0].set_title('Random Oversampling')
sns.scatterplot(x=X_smote[:, 1], y=X_smote[:, 2], hue=y_smote, ax=axes[1][1])
axes[1][1].set_title('SMOTE')
plt.show()
```
