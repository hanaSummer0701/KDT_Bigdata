# 5. 데이터 타입 변경

```python
# 데이터 프레임 생성

# 날짜 데이터
dates = pd.date_range('2024-01-01', periods=5)

# 숫자 데이터
numbers = [10, 20, 30, 40, 50]

# 범주형 데이터
categories = ['A', 'B', 'C', 'A', 'B']

# 문자열 데이터
strings = ['apple', 'banana', 'orange', 'grape', 'melon']

# 데이터프레임 생성
df = pd.DataFrame({
    'Dates': dates,
    'Numbers': numbers,
    'Categories': categories,
    'Strings': strings
})

# 데이터프레임 정보 출력
print("DataFrame 정보:")
print(df.info())

# 데이터프레임 샘플 출력
print("\nDataFrame 샘플:")
print(df.head())
```

```python
# datetime 형태로 변하는 코드
df['Dates'] = pd.to_datetime(df['Dates'])
df
```

```python
df.dtypes
```

```python
# 날짜 간 시간 차 계산
df['date_diff'] = df['Dates'].diff().dt.days
df
```

```python
# 연, 월 추출
df['year'] = df['Dates'].dt.year
df['month'] = df['Dates'].dt.month
df
```

```python
df.dtypes
```

```python
# date_diff 결측치 제거 후 int 형태로 변환
df.isna().sum()
```

```python
# 0으로 결측치 채움
df['date_diff']=df['date_diff'].fillna(0)
df
```

```python
# 타입 변환
df['date_diff']=df['date_diff'].astype(int)
df.dtypes
```

# 6. 시각화

```python
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

sns.set(font_scale=1.4)
# 맑은 고딕으로 글꼴 설정하기
plt.rc('font', family='Malgun Gothic') 
# 마이너스 기호 문제 해결하기
mpl.rcParams['axes.unicode_minus'] = False
```

```python
house.dtypes
```

```python
# 히스토그램 그리기
plt.figure(figsize=(12, 8))
for column in house.columns:
    plt.hist(house[column], bins=20, alpha=0.5, label=column)

# 그래프 제목 및 라벨 설정
plt.title('Distribution of Features', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend()

# 그래프 출력
plt.show()
```

```python
# line plot
# 라인 플롯 그리기
plt.figure(figsize=(10, 6))

# median_income 라인 플롯
plt.plot(house['median_income'], label='median_income', color='blue')

# median_house_value 라인 플롯
plt.plot(house['median_house_value'], label='median_house_value', color='orange')

# 그래프 제목 및 라벨 설정
plt.title('Trends of Median Income and Median House Value', fontsize=14)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend()

# 그래프 출력
plt.show()
```

```python
# regplot 그리기
plt.figure(figsize=(10, 6))
sns.regplot(x='median_income', y='median_house_value', data=house)

# 그래프 제목 설정
plt.title('Regression Plot of Median Income vs Median House Value', fontsize=14)
plt.xlabel('Median Income', fontsize=12)
plt.ylabel('Median House Value', fontsize=12)

# 그래프 출력
plt.show()
```
