#### 그 외에도 컬럼 바꾸는 방법

```python
# Custom Function 생성 후 값 변경

def change_value(values):
    if values == 'Portuguese':
        return 1
    elif values == 'German':
        return 2
    elif values == 'Spanish':
        return 3
    elif values == 'Italian':
        return 4
    elif values == 'Dutch':
        return 5
    elif values == 'English':
        return 6
    elif values == 'Lithuanian':
        return 7
    elif values == 'Angolan':
        return 8
    elif values == 'Cape Verdean':
        return 9
    elif values == 'Guinean':
        return 10
    elif values == 'Mozambican':
        return 11
    elif values == 'Santomean':
        return 12
    elif values == 'Turkish':
        return 13
    elif values == 'Brazilian':
        return 14
    elif values == 'Romanian':
        return 15
    elif values == 'Moldova':
        return 16
    elif values == 'Mexican':
        return 17
    elif values == 'Ukrainian':
        return 18
    elif values == 'Russian':
        return 19
    elif values == 'Cuban':
        return 20
    elif values == 'Colombian':
        return 21
# 함수 적용 (apply)
dropout['Nacionality'] = dropout['Nacionality'].apply(change_value)
```

```python
# scikit-learn 라이브러리 사용

# label encoding
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(dropout['Nacionality'])
dropout['Nacionality'] = encoder.transform(dropout['Nacionality'])
```

```python
# pd.get_dummies 사용

# train 데이터셋의 범주 개수와 test 데이터셋의 범주 개수가 다른 경우 사용 X -> OnehotEncoder 등으로 대체
pd.get_dummies(dropout['Nacionality'], dummy_na=True)
```

```python
# 데이터 컬럼 추가

# 간단한 방법 - 무조건 맨 오른쪽에 위치. 원하는 위치로 설정 어렵
dropout['Nationality'] = 999

# insert 사용 - 원하는 위치에 넣기 가능
# insert(loc = “넣고 싶은 새로운 컬럼의 위치”, columns = “추가하고 싶은 새로운 컬럼명”, value = “새로운 컬럼의 값”)
dropout.insert(loc=2, column='Nationality', value=999)
```

```python
# 데이터프레임.drop(columns = {“지우고 싶은 컬럼명”, “지우고 싶은 컬럼명”}
dropout = dropout.drop(columns={'Nacionality'})
```

### 2-4. 데이터 결측치 처리(대치)

1. 최빈값 : 범주형 변수의 경우 (가장 비율이 높은 범주로 대치 가능)
2. 중앙값 : 연속형 변수의 경우 중앙값으로 대치 가능
3. 평균 : 연속형 변수의 경우 평균값으로 대치 가능
4. 임의 지정 : 분석가가 원하는 수치로 대치 가능

```python
# float 타입 컬럼 결측치 제거
 # 한 값이 많음 -> 최빈
 # 값이 여러 개, 한 값만 많지 않음 -> 평균
```

```python
# 최빈값

columns_fill_max = ['Application order', 'Curricular units 1st sem (credited)','Curricular units 1st sem (without evaluations)','Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (without evaluations)']

for col in columns_fill_max:
    max_val = dropout2[col].value_counts().idxmax()
    dropout2[col] = dropout2[col].fillna(max_val)
```

```python
# 평균

columns_fill_means = ['Age at enrollment', 'Curricular units 1st sem (enrolled)', 'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)'
                      , 'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (enrolled)', 'Curricular units 2nd sem (evaluations)',  'Curricular units 2nd sem (approved)'
                      , 'Curricular units 2nd sem (grade)',  'Unemployment rate', 'Inflation rate', 'GDP']

for col in columns_fill_means:
    means = dropout2[col].mean()
    dropout2[col] = dropout2[col].fillna(means)
```

```python
# object 타입 컬럼 결측치 제거
 # 최빈
```

```python
for col in object_list:
    max_val = dropout2[col].value_counts().idxmax()
    dropout2[col] = dropout2[col].fillna(max_val)
```
