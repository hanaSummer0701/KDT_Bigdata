```python
def chk_cat(x):
    if '더빙'in x:
        return '더빙'
    elif any([i in x for i in['뉴스', '라디오', '라이브커머스']]):
        return '컨텐츠'
    elif '크로마키' in x:
        return '크로마키'
    else:
        return x
```

```python
media['강좌내용']=media['강좌내용'].apply(lambda x: chk_cat(x))
```
