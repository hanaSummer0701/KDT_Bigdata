```python
# read the mask image
# taken from
# http://www.stencilry.org/stencils/movies/alice%20in%20wonderland/255fk.jpg
alice_mask = np.array(Image.open("sphx_glr_masked_002.png"))
```

```python
np.array(Image.open("sphx_glr_masked_002.png"))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")
```

```python
STOPWORDS # 불용어사전:해당 단어를 분석할 대상으로 취급하지 않겠다!라고 하는 것을 모아 둔 단어.
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_mask,
               stopwords=stopwords, contour_width=3, contour_color='steelblue')

#  max_words=2000: 2000개 단어만 보이게 하겠다.
# mask: 워드클라우드 형태 현재는 alice
# contour_width: 외곽선 굵기. 안 적으면 외곽선 x
# contour_color='steelblue': steelblue 색상으로 외곽선 그리기
```

```python
# generate word cloud -> 워드클라우드 생성
wc.generate(text3)
```

```python
# store to file # 파일 저장
wc.to_file("alice_mask.png")

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from wordcloud import ImageColorGenerator
```

```python
# read the mask / color image taken from
# http://jirkavinse.deviantart.com/art/quot-Real-Life-quot-Alice-282261010
alice_coloring = np.array(Image.open("alice_color.jpg"))
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)
# generate word cloud
wc.generate(text4)
```

```python
# create coloring from image
image_colors = ImageColorGenerator(alice_coloring)
```

```python
plt.imshow(wc.recolor(color_func=image_colors))
plt.axis("off")
plt.show()
```

```python
# show
fig, axes = plt.subplots(1, 3)
# (1,3): 단순마스킹, 컬러마스킹, 원본 이미지
axes[0].imshow(wc, interpolation="bilinear")
# recolor wordcloud and show
# we could also give color_func=image_colors directly in the constructor
axes[1].imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
axes[2].imshow(alice_coloring, cmap=plt.cm.gray, interpolation="bilinear")
for ax in axes:
    ax.set_axis_off()
plt.show()
```
